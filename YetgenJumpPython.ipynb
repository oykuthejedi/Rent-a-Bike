```python
import datetime

class BikeRent:
    
    def __init__(self,stock=0):
        

        self.stock = stock

    def displaystock(self):
        

        print("Şu anda kiralanabilecek {} tane bisikletimiz vardır.".format(self.stock))
        return self.stock

    def rentBikeHourly(self, n):
       
        if n <= 0:

            print("Bisiklet Sayımız Pozitif Olmalıdır.")
            return None
        
        elif n > self.stock:

            print("Malesef şu an kiralanabilecek bisiklet sayımız {} tanedir.".format(self.stock))
            return None
        
        else:
            now = datetime.datetime.now()                      
            print("{} adet bisikleti {} saat kiraladınız".format(n,now.hour))
            print("Bisiklet başına her saat için 7 TL ödemeniz gerekmektedir.")
            print("İyi Günler.")

            self.stock -= n
            return now      
     
    def rentBikeDaily(self, n):
        
        if n <= 0:

            print("Bisiklet Sayımız Pozitif Olmalıdır.")
            return None

        elif n > self.stock:

            print("Malesef şu an kiralanabilecek bisiklet sayımız {} tanedir.".format(self.stock))
            return None
    
        else:

            now = datetime.datetime.now()                      
            print("{} adet bisikleti {} gün kiraladınız".format(n, now.hour))
            print("Bisiklet başına her gün için 50 TL ödemeniz gerekmektedir.")
            print("İyi Günler.")

            self.stock -= n
            return now
        
    def rentBikeWeekly(self, n):
        
        if n <= 0:

            print("Bisiklet Sayımız Pozitif Olmalıdır.")
            return None

        elif n > self.stock:
          
            print("Malesef şu an kiralanabilecek bisiklet sayımız {} tanedir.".format(self.stock))
            return None        
        
        else:
            now = datetime.datetime.now()
            print("{} adet bisikleti {} hafta kiraladınız".format(n, now.hour))
            print("Bisiklet başına her hafta için 125 TL ödemeniz gerekmektedir.")
            print("İyi Günler.")
            self.stock -= n

            return now
    

    
    def returnBike(self, request):
        """
        1. Accept a rented bike from a customer
        2. Replensihes the inventory
        3. Return a bill
        """
        rentalTime, rentalBasis, numOfBikes = request
        bill = 0

        if rentalTime and rentalBasis and numOfBikes:
            self.stock += numOfBikes
            now = datetime.datetime.now()
            rentalPeriod = now - rentalTime
        
            # hourly bill calculation
            if rentalBasis == 1:
                bill = round(rentalPeriod.seconds / 3600) * 7 * numOfBikes
                
            # daily bill calculation
            elif rentalBasis == 2:
                bill = round(rentalPeriod.days) * 50 * numOfBikes
                
            # weekly bill calculation
            elif rentalBasis == 3:
                bill = round(rentalPeriod.days / 7) * 125 * numOfBikes
            
               
            if (3 <= numOfBikes <= 5):
                print("%30'luk İndiriminiz Uygulanmıştır.")
                bill = bill * 0.7

            print("Bizi Tercih Ettiğiniz İçin Teşekkür Ederiz.")
            print("Ödemeniz Gereken Ücret {}".format(bill))
            return bill
        else:
            print("Bisiklet Kiraladığınızdan Emin Misiniz?")
            return None



class Customer:

    def __init__(self):
        
        
        self.bikes = 0
        self.rentalBasis = 0
        self.rentalTime = 0
        self.bill = 0

    
    def requestBike(self):
        
                      
        bikes = input("Kaç Tane Bisiklet Kiralamak İstiyorsunuz ?")
        try:
            bikes = int(bikes)
        except ValueError:
            print("Pozitif Bir Tam Sayı Değil.")
            return -1

        if bikes < 1:
            print("Bisiklet Sayısı Sıfırdan Büyük Olmalıdır.")
            return -1
        else:
            self.bikes = bikes
        return self.bikes
     
    def returnBike(self):
        
        if self.rentalBasis and self.rentalTime and self.bikes:
            return self.rentalTime, self.rentalBasis, self.bikes  
        else:
            return 0,0,0


def main():
    shop = BikeRent(100)
    customer = Customer()

    while True:
        print("""
        ====== Bike Rental Shop =======
        1. Mevcut Bisiklet Sayısını Görüntüleme
        2. Saatlik Ödemeniz Gereken Ücret 7 TL.
        3. Günlük Ödemeniz Gereken Ücret 50 TL.
        4. Haftalık Ödemeniz Gereken Ücret 125 TL.
        5. Bisikleti İade Et.
        6. Çıkış
        """)
        
        choice = input("Seçiminiz: ")
        
        try:
            choice = int(choice)
        except ValueError:
            print("Hatalı Seçim")
            continue
        
        if choice == 1:
            shop.displaystock()
        
        elif choice == 2:
            customer.rentalTime = shop.rentBikeHourly(customer.requestBike())
            customer.rentalBasis = 1

        elif choice == 3:
            customer.rentalTime = shop.rentBikeDaily(customer.requestBike())
            customer.rentalBasis = 2

        elif choice == 4:
            customer.rentalTime = shop.rentBikeWeekly(customer.requestBike())
            customer.rentalBasis = 3

        elif choice == 5:
            customer.bill = shop.returnBike(customer.returnBike())
            customer.rentalBasis, customer.rentalTime, customer.bikes = 0,0,0        
        elif choice == 6:
            break
        else:
            print("Lütfen 1-6 arasında bir sayı girin")        
    print("Bizi Tercih Ettiğiniz İçin Teşekkür Ederiz.")


if __name__=="__main__":
    main()
```
